```python
nest_asyncio.apply()  # Allows nested async loops (for Jupyter)

async def scrape_expedia():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)  # Set to True once tested
        page = await browser.new_page()
        url = 'https://www.expedia.co.uk/Hotel-Search?destination=Antalya%2C%20Antalya%20Region%2C%20T%C3%BCrkiye&regionId=481&latLong=36.886208%2C30.70644&flexibility=0_DAY&d1=2025-06-23&startDate=2025-06-23&d2=2025-06-29&endDate=2025-06-29&adults=2%2C2&rooms=2&isInvalidatedDate=false&upsellingNumNightsAdded=&theme=&userIntent=&semdtl=&upsellingDiscountTypeAdded=&useRewards=false&sort=RECOMMENDED'

        await page.goto(url)
        await page.wait_for_timeout(5000)  # 5 sec wait for JS content
        content = await page.content()
        await browser.close()
        soup = BeautifulSoup(content, 'html.parser')
        html = soup.prettify()
        print(html)

        text_maker = html2text.HTML2Text()
        text_maker.ignore_links = False  # Keep links
        text_maker.body_width = 0        # Don't wrap lines
        
        expedia_text = text_maker.handle(html)
        with open("expedia.txt", "a") as f:
            f.write(expedia_text)

# Run the coroutine safely
await scrape_expedia()
```

```python
import urllib.parse
from datetime import date, timedelta

def build_booking_url(location, checkin, checkout, adults=2, children=0, rooms=1):
    base_url = "https://www.booking.com/searchresults.en-gb.html"
    query = {
        "ss": location,
        "checkin": checkin.strftime("%Y-%m-%d"),
        "checkout": checkout.strftime("%Y-%m-%d"),
        "group_adults": adults,
        "group_children": children,
        "no_rooms": rooms,
        "nflt": "ht_id=204;review_score=60"
    }
    return f"{base_url}?{urllib.parse.urlencode(query)}"

# Example usage
checkin = date(2025, 5, 12)
checkout = checkin + timedelta(days=1)

url = build_booking_url("skdbvjrkbv", checkin, checkout)
print(url)
```
