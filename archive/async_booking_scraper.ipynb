```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup
import html2text
import numpy as np
import pandas as pd
from scipy import stats
import nest_asyncio
from playwright.async_api import async_playwright
from playwright_stealth import stealth_async
from urllib.parse import urlparse, urlencode, parse_qs
from datetime import datetime, timedelta
import time
import re
from fuzzywuzzy import process
```

```python
nest_asyncio.apply()  # Allows nested async loops (for Jupyter)
HEADERS = {'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36',
           'Accept-Language': 'en-US, en;q=0.5'
          }
semaphore = asyncio.Semaphore(10)

async def build_url(location, from_, to_, sort, adults=2, children=0, rooms=1, holiday_length=4):
    # Convert checkin string to datetime
    checkin_date = datetime.strptime(from_, "%Y-%m-%d")
    
    # Get initial checkout date
    checkout_date = checkin_date + timedelta(days=holiday_length)
    
    # Get datetime for end of window
    end_date = datetime.strptime(to_, "%Y-%m-%d")
    
    # Calculate window range
    window = (end_date - checkin_date).days
    range_ = window + 1 - holiday_length
    
    order_map = {'price': ['price'], 'rating': ['bayesian_review_score'],
                 'price & rating': ['price', 'bayesian_review_score']}

    urls = []
    dates = []
    for map_ in range(len(order_map[sort])):
        for i in range(range_):
            checkin_new_date = checkin_date + timedelta(days=i)
            checkout_new_date = checkout_date + timedelta(days=i)
            
            # Convert back to string
            checkin_new = checkin_new_date.strftime("%Y-%m-%d")
            checkout_new = checkout_new_date.strftime("%Y-%m-%d")

            # Create url
            base_url = "https://www.booking.com/searchresults.en-gb.html"
            query = {
                "ss": location,
                "checkin": checkin_new,
                "checkout": checkout_new,
                "group_adults": adults,
                "group_children": children,
                "no_rooms": rooms,
                "order": order_map[sort][map_],
                "nflt": "ht_id=204;review_score=60"
            }
            url = f"{base_url}?{urlencode(query)}"
    
            urls.append(url)
            dates.append([checkin_new, checkout_new])
    
    return urls, dates


# Asynchronous fetch
async def fetch(session, url):
    async with semaphore:
        try:
            async with session.get(url, timeout=10) as response:
                return await response.text()
        except Exception as e:
            print(f"Error fetching {url}: {e}")
            return None

async def extract_hotels_from_page(html, date):
    hotels_data = {'name': [], 'location': [], 'checkin date': [], 'checkout date': [],
                   'link': [], 'price': [], 'rating': [], 'reviews': []}

    page = BeautifulSoup(html, 'lxml') 
    hotels = page.findAll('div', {'data-testid': 'property-card'})
    links = page.findAll('a', {'data-testid': 'availability-cta-btn'}, href=True)
    locations = page.findAll('span', {'data-testid': 'address'})
    hotels = [h.text.strip() for h in hotels]

    # Loop over the hotel elements and extract the desired data
    for hotel, a, loc in zip(hotels, links, locations):
        try:
            name = hotel.split('Opens in new window')[0].strip()
            link = a['href']
            location = ', '.join(loc)
            rating = float(hotel.split('Scored')[1][:4].strip())
            review = int(re.sub(r'[^0-9]', '', hotel.split('reviews')[0][-8:]))
            try:
                price = int(re.sub(r'[^0-9]', '', hotel.split('Price')[1][:8]))
            except:
                price = int(re.sub(r'[^0-9]', '', hotel.split('Current price')[1][:8]))
    
            hotels_data['name'].append(name)
            hotels_data['location'].append(location)
            hotels_data['checkin date'].append(date[0])
            hotels_data['checkout date'].append(date[1])
            hotels_data['link'].append(link)
            hotels_data['price'].append(price)
            hotels_data['rating'].append(rating)
            hotels_data['reviews'].append(review)
        except:
            continue

    hotels_df = pd.DataFrame(hotels_data)
    
    return hotels_df


# Master coroutine: fetch and parse a list of URLs
async def booking_search(location, from_, to_, adults=2, children=0, rooms=1, sort='price', holiday_length=4):
    connector = aiohttp.TCPConnector(limit=10)
    async with aiohttp.ClientSession(headers=HEADERS, connector=connector) as session:
        urls, dates = await build_url(location, from_, to_, sort, adults, children, rooms, holiday_length)
        
        tasks = [fetch(session, url) for url in urls]
        html_pages = await asyncio.gather(*tasks)

        tasks = [extract_hotels_from_page(html, date) for html, date in zip(html_pages, dates)]
        hotels_list = await asyncio.gather(*tasks)


    all_best_hotels = pd.concat(hotels_list)

    # Scale columns
    all_best_hotels['price_percentile'] = stats.percentileofscore(all_best_hotels['price'], all_best_hotels['price'])*0.01
    all_best_hotels['rating_scaled'] = all_best_hotels['rating']*0.1
    
    # Build VM (Value for Money) score
    all_best_hotels['vm_score_unrounded'] = 100*(((1-all_best_hotels['price_percentile'])+all_best_hotels['rating_scaled'])/2)
    
    
    if sort == 'price & rating':
        all_best_hotels = all_best_hotels.sort_values(by='vm_score_unrounded', ascending=False)
        
    elif sort == 'price':
        all_best_hotels = all_best_hotels.sort_values(by='price', ascending=True)
    
    else:
        all_best_hotels = all_best_hotels.sort_values(by='rating', ascending=False)

    # Apply rounding to VM score
    all_best_hotels['vm_score'] = round(all_best_hotels['vm_score_unrounded'])
    all_best_hotels['vm_score'] = all_best_hotels['vm_score'].astype(int)

    # Drop scaled columns
    all_best_hotels.drop('price_percentile', axis=1, inplace=True)
    all_best_hotels.drop('rating_scaled', axis=1, inplace=True)
    all_best_hotels.drop('vm_score_unrounded', axis=1, inplace=True)
    
    # Drop duplicates
    all_best_hotels = all_best_hotels.drop_duplicates(subset=['name', 'rating', 'reviews',
                                          'checkin date', 'checkout date'])
    
    return all_best_hotels
```

```python
semaphore = asyncio.Semaphore(5)
async def build_booking_flight_url(airport_from, airport_to, adults, children, from_, to_, holiday_length):
    iata_table = pd.read_csv('airports.csv')

    # Convert checkin string to datetime
    depart_date = datetime.strptime(from_, "%Y-%m-%d")
    # print(depart_date)
    
    # Get initial checkout date
    return_date = depart_date + timedelta(days=holiday_length)
    
    # Get datetime for end of window
    end_date = datetime.strptime(to_, "%Y-%m-%d")
    
    # Calculate window range
    window = (end_date - depart_date).days
    range_ = window + 1 - holiday_length
    
    base_url = 'https://flights.booking.com/flights/FROM.AIRPORT-TO.AIRPORT/'
    airport_from_clean = process.extractOne(airport_from, iata_table['Airport name'])[0]
    airport_to_clean = process.extractOne(airport_to, iata_table['Airport name'])[0]
    from_code = iata_table['IATA'][iata_table['Airport name'] == airport_from_clean].values[0]
    to_code = iata_table['IATA'][iata_table['Airport name'] == airport_to_clean].values[0]
    base_url = base_url.replace('FROM', from_code)
    base_url = base_url.replace('TO', to_code)

    
    urls = []
    dates = []
    for i in range(range_):
        depart_new_date = depart_date + timedelta(days=i)
        return_new_date = return_date + timedelta(days=i)
        
        # Convert back to string
        depart_new = depart_new_date.strftime("%Y-%m-%d")
        return_new = return_new_date.strftime("%Y-%m-%d")
        
        query = {
            "type": "ROUNDTRIP",
            "cabinClass": "ECONOMY",
            "adults": adults,
            "children": children,
            "from": from_code + ".AIRPORT",
            "to": to_code + ".AIRPORT",
            "depart": depart_new,
            "return": return_new,
            "sort": "CHEAPEST",
            "travelPurpose": "leisure"
        }
        url = f"{base_url}?{urlencode(query)}"
        
        urls.append(url)
        dates.append([depart_new, return_new])

    return urls, dates


async def get_booking_flight_prices(context, url, daterange):
    flights_data = {'depart_date': [], 'return_date': [], 'avg_price': [], 'source': []}
    try:
        page = await context.new_page()
        # page.set_default_timeout(60000)
        await page.set_viewport_size({"width": 800, "height": 600})
        await page.route("**/*", lambda route: route.abort() 
                         if route.request.resource_type in ["image", "font", "stylesheet"] else route.continue_())
        await page.goto(url, wait_until="domcontentloaded", timeout=60000)
        await page.wait_for_selector('[data-testid="flight_card_price_total_price"]', timeout=60000)
        prices_text = await page.locator('[data-testid="flight_card_price_total_price"]').all_inner_texts()
        await page.evaluate("""() => {document.body.innerHTML = "";}""")
        prices = [float(p.split(' ')[0][1:].replace(',', '')) for p in prices_text]
        avg_price = round(np.mean(prices), 2)
    
        flights_data['depart_date'].append(daterange[0])
        flights_data['return_date'].append(daterange[1])
        flights_data['avg_price'].append(avg_price)
        flights_data['source'].append(url)
            
    except:
        flights_data['depart_date'].append(daterange[0])
        flights_data['return_date'].append(daterange[1])
        flights_data['avg_price'].append(99999)
        flights_data['source'].append(url)
    finally:
        await page.close()

    flights_df = pd.DataFrame(flights_data)

    return flights_df


async def booking_flights_search(airport_from, airport_to, adults, children, from_, to_, holiday_length):
    urls, dates = await build_booking_flight_url(airport_from, airport_to, adults, children, from_, to_, holiday_length)
    
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)

        async def task(url, daterange):
            async with semaphore:
                context = await browser.new_context()
                result = await get_booking_flight_prices(context, url, daterange)
                await context.close()
                return result

        tasks = [task(url, daterange) for url, daterange in zip(urls, dates)]
        flights_list = await asyncio.gather(*tasks)
        await browser.close()

    all_flight_info = pd.concat(flights_list)
    all_flight_info = all_flight_info.sort_values(by='avg_price', ascending=True)
    
    return all_flight_info
```

```python
nest_asyncio.apply()
async def build_kayak_flight_url(airport_from, airport_to, adults, children, from_, to_, holiday_length):
    iata_table = pd.read_csv('airports.csv')

    airport_from_clean = process.extractOne(airport_from, iata_table['Airport name'])[0]
    airport_to_clean = process.extractOne(airport_to, iata_table['Airport name'])[0]
    from_code = iata_table['IATA'][iata_table['Airport name'] == airport_from_clean].values[0]
    to_code = iata_table['IATA'][iata_table['Airport name'] == airport_to_clean].values[0]

    adults_str = f'{adults}adults/'
    
    if children == 0:
        children_str = ''
    else:
        add_list = ['11' for i in range(children)]
        add_str = '-'.join(add_list)
        children_str = 'children-' + add_str

    sort_str = '?sort=bestflight_a'

    # Convert checkin string to datetime
    depart_date = datetime.strptime(from_, "%Y-%m-%d")

    # Get datetime for end of window
    end_date = datetime.strptime(to_, "%Y-%m-%d")
    
    # Calculate window range
    window = (end_date - depart_date).days

    num_searches = int(window/32) + 1

    urls = []
    dates = []
    for n in range(num_searches):
        depart_new_dt = depart_date + timedelta(days=n*32)
        month_later_dt = depart_new_dt + timedelta(days=32)
        depart_new = depart_new_dt.strftime("%Y-%m-%d")
        month_later = month_later_dt.strftime("%Y-%m-%d")
    
        base_url = f'https://www.kayak.co.uk/flights/{from_code}-{to_code}/{depart_new}/{month_later}-flexible-calendar-{holiday_length}/'
        url = base_url + adults_str + children_str + sort_str
        urls.append(url)
        dates.append([depart_new, month_later])

    return urls, dates, airport_from_clean, airport_to_clean, end_date


async def get_kayak_flight_prices(url, daterange, holiday_length, end_date):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)  # headless=False to avoid detection
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
            viewport={"width": 1280, "height": 800}
        )

        page = await context.new_page()
        await stealth_async(page)
        await page.goto(url, wait_until="domcontentloaded", timeout=60000)
        # await page.wait_for_selector('[class="nSfL-price"]', timeout=60000)

        dates_text = await page.locator('[class="nSfL-date"]').all_inner_texts()
        prices_text = await page.locator('[class="nSfL-price"]').all_inner_texts()
        # print(dates_text)
        # print(prices_text)

        await browser.close()

        flight_page_url = url.replace(f'-flexible-calendar-{holiday_length}', '')
        depart_date = datetime.strptime(daterange[0], "%Y-%m-%d")
        day_start = str(int(daterange[0].split('-')[-1]))
        start_from_index = dates_text.index(day_start)
        dates_text = dates_text[start_from_index:]
        prices_text = prices_text[start_from_index:]
        

        flights_data = {'depart_date': [], 'return_date': [], 'price': [], 'url': []}
        for i in range(len(dates_text)):
            depart_update_dt = depart_date + timedelta(days=i)
            return_update_dt = depart_date + timedelta(days=i+holiday_length)
            depart_update = depart_update_dt.strftime("%Y-%m-%d")
            return_update = return_update_dt.strftime("%Y-%m-%d")
            date_url = flight_page_url.replace(daterange[0], depart_update)
            date_url = date_url.replace(daterange[1], return_update)
            
            try:
                price = int(re.sub(r'[^0-9]', '', prices_text[i]))
            except:
                price = 9999

            flights_data['depart_date'].append(depart_update)
            flights_data['return_date'].append(return_update)
            flights_data['price'].append(price)
            flights_data['url'].append(date_url)
        
        flights_df = pd.DataFrame(flights_data)

        if end_date < return_update_dt:
            end = end_date.strftime("%Y-%m-%d")
            last_date = list(flights_df.index[flights_df['return_date'] == end])[0]
            flights_df = flights_df.iloc[:last_date + 1].copy()
            
        return flights_df

async def kayak_flights_search(airport_from, airport_to, adults, children, from_, to_, holiday_length):
    urls, dates, airport_from_clean, airport_to_clean, end_date = await build_kayak_flight_url(airport_from, airport_to, adults, children, from_, to_, holiday_length)
    tasks = [get_kayak_flight_prices(url, daterange, holiday_length, end_date) for url, daterange in zip(urls, dates)]
    flights_list = await asyncio.gather(*tasks)
    
    all_flight_info = pd.concat(flights_list)
    all_flight_info['airport_from'] = airport_from_clean
    all_flight_info['airport_to'] = airport_to_clean
    all_flight_info = all_flight_info[['airport_from', 'airport_to', 'depart_date', 'return_date', 'price', 'url']].copy()
    all_flight_info['price'] = all_flight_info['price']*adults*children
    all_flight_info = all_flight_info.sort_values(by='price', ascending=True)
    all_flight_info = all_flight_info.replace(9999*adults*children, 'Missing')
    
    return all_flight_info
```
