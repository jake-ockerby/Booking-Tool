```python
from booking_tools import Booker
import asyncio
import time
```

```python
booker = Booker(location='Milan', from_='2025-07-20', to_='2025-08-01', adults=2, children=0,
               rooms=1, sort='Price', holiday_length=4, airport_from='London',
               airport_to='Milan Malpensa', review_score=8, mealplan='Breakfast', twin_beds='Twin Beds',
               stars=4, distance='1 Mile')
```

```python
booker = Booker(location='Antalya', from_='2025-06-11', to_='2025-07-11', adults=2, children=0,
               rooms=1, sort='Price', holiday_length=7)
```

```python
start_time = time.time()
hotels_flights_df = asyncio.run(booker.booking_search())
end_time = time.time()

print('Execution Time: '+ str(round(end_time - start_time)) + 's')
```
